# 실습 내용

- Random Search로 KNN 알고리즘 모델을 튜닝한다.

# 1.환경 준비

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'data/mobile_cust_churn.csv'
df = pd.read_csv(path)
```

# 2.데이터 이해

```python
# 상위 몇 개 행 확인
df.head()
```

```python
# 기술통계 확인
df.describe()
```

# 3.데이터 전처리

**1) 변수 제거**

```python
# 제거 대상: id
drop_cols = ['id']

# 변수 제거
df.drop(drop_cols, axis=1, inplace=True)

# 확인
df.head()
```

**2) x, y 분리**

```python
# Target 설정
target = 'CHURN'

# 데이터 분리
X = df.drop(target, axis=1)
y = df[target]
```

**3) 가변수화**

```python
# 가변수화 대상: REPORTED_SATISFACTION, REPORTED_USAGE_LEVEL, CONSIDERING_CHANGE_OF_PLAN
df_dummies = ['REPORTED_SATISFACTION', 'REPORTED_USAGE_LEVEL', 'CONSIDERING_CHANGE_OF_PLAN']

# 가변수화
X = pd.get_dummies(X, columns=df_dummies, drop_first=True)

# 확인
X.head()
```

**4) 학습용, 평가용 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.7,
                                                    random_state=1
                                                    )
```

5) 정규화

```python
# 모듈 불러오기
from sklearn.preprocessing import MinMaxScaler

# 정규화
scaler = MinMaxScaler()
scaler.fit(X_train, y_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# 4.성능 예측

- k-Fold Cross Validation을 사용해 모델의 성능을 예측한다.

```python
# 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
```

```python
# 선언하기
model_dt = DecisionTreeClassifier(random_state=1)
```

```python
# 성능예측
cv_score = cross_val_score(model_dt, X_train, y_train, cv=5, scoring='accuracy')
```

```python
# 결과확인
print(cv_score)
print(cv_score.mean())
```

# 5.모델 튜닝


- Random Search로 튜닝을 진행한다.

**1) 모델 튜닝**

```python
# 파라미터 선언
  # n_neighbors: range(3, 31)
param = {'max_depth': range(3,31)}

# Random Search 선언
  # cv=5
  # n_iter=20
  # scoring='accuracy'

model = RandomizedSearchCV(model_dt, param, cv=5, n_iter=20, scoring='accuracy')
```

```python
# 학습하기
model.fit(X_train, y_train)
```

**2) 결과 확인**

```python
# mean_test_score 확인
print(model.cv_results_['mean_test_score'])
print('-'*70)
print('최적의 파라미터 :', model.best_params_)
print('-'*70)
print('최고의 성능 :', model.best_score_)
```

```python
plt.figure(figsize=(8,8))
plt.barh(y=list(X),width=model.best_estimator_.feature_importances_)
plt.show()
```

# 6.성능 평가

- 학습을 통해 예상한 성능과 실제 평가에 따른 성능은 차이가 있을 수 있다.


```python
# 예측하기
y_pred = model.predict(X_test)
```

```python
# 평가하기
print('cm :', confusion_matrix(y_test, y_pred))
print('cfc :', classification_report(y_test, y_pred))
```
