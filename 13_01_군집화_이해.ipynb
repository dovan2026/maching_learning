```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.scatterplot(x='var_1', y='var_2', data=df)
plt.show()
```

# 2.군집화 이해

**1) 라이브러리 불러오기**

```python
from sklearn.cluster import KMeans
```

**2) 선언하기**

```python
model_kmeans = KMeans(n_clusters=3, random_state=1)
```

**3) 학습하기**

```python
model_kmeans.fit(df)
```

**4) 예측하기**

```python
model_kmeans.predict(df)
```

```python
df['label'] = model_kmeans.predict(df)
df.head()
```

```python
sns.scatterplot(x='var_1',
                y='var_2',
                data=df,
                hue='label',
                palette='rainbow')
```

```python
# 학습된 모델의 inertia 값을 확인(군집 내 데이터와 중심점 간 거리의 제곱합)
# 값이 작을수록 군집 내 데이터가 군집 중심에 더 가까이 모여 있음
# 값이 작을수록 군집 내 점들이 중심에 더 빽뺵하게 모여 있다는 뜻(응집도가 증가)

model_kmeans.inertia_
```

- 군집 개수 늘려보기

```python
temp_model = KMeans(n_clusters=500, random_state=1)
temp_model.fit(df)
```

```python
# 같은 데이터에서 같은 전처리 조건이라면 inertia_ 값이 작을 수록 좋다고 볼 수 있으나, 무조건 작을수록 좋은 건 아님
# 군집 개수 k를 늘리면 inertia는 거의 항상 줄어듭니다.
# 극단적으로 데이터 300개를 군집 300개로 나누면 각 데이터가 자기 자신만의 군집이 됩니다. 그러면 중심점과의 거리가 거의 0이 됨.
temp_model.inertia_
```

- 최적의 군집 개수 찾기

```python
distance = []

for k in range(2,10):                   # k값을 바꿔가며 최적의 k값 찾기
    k_model = KMeans(n_clusters=k)
    k_model.fit(df)
    distance.append(k_model.inertia_)   # 각 k에 대한 inertia 값 저장
```

```python
# 시각화 : Elbow method
# 좋은 k 후보 = inertia가 충분히 작아지면서도 군집 개수가 너무 많지 않은 지점.
sns.lineplot(x=range(2,10), y=distance)
```
