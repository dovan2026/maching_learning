# 실습 내용

- K-Fold Cross Validation을 사용해 모델의 성능을 예측한다.

# 1.환경 준비

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'data/boston.csv'
df = pd.read_csv(path)
```

# 2.데이터 이해

```python
# 데이터 살펴보기
df.head()
```

```python
# 기술통계 확인
df.describe()
```

# 3.데이터 전처리

**1) x, y 분리**

```python
# Target 확인
target = 'medv'

# x, y 분리
X = df.drop(target, axis=1)
y = df[target]
```

**2) 학습용, 평가용 데이터 분리**

```python
# 라이브러리 불러오기
from sklearn.model_selection import train_test_split

# 학습용, 평가용 데이터 7:3으로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state=1                                                
                                                    )
```

**3) 정규화**

```python
# 모듈 불러오기
from sklearn.preprocessing import MinMaxScaler

# 정규화
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

# 4.성능 예측

- K분할 교차 검증 방법으로 모델 성능을 예측
- `cross_val_score(model, x_train, y_train, cv=n)` 형태로 사용
- cv 옵션에 k값(분할 개수, 기본값=5)을 지정
- cross_val_score 함수는 넘파이 배열 형태의 값을 반환
- cross_val_score 함수 반환 값의 평균을 해당 모델의 예측 성능으로 볼 수 있다.


**1) Linear Regression**

```python
# 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# 선언하기
model_lr = LinearRegression()

# 검증하기
cv_score = cross_val_score(model_lr, X_train_s, y_train, cv=10, scoring='r2')

# 확인
print(cv_score)
print('평균: ', cv_score.mean())            # 모델의 평균 성능 평가(일반적인 성능을 대표하는 값)
print('표준편차: ', cv_score.std())          # 모델 성능의 일관성 평가(표준편차가 작다면 -> 모델이 안정적)

# 성능 정보 수집
result = {}
result['lr'] = cv_score.mean()
```

**2) KNN**

```python
# 불러오기
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score

# 선언하기
model_knn = KNeighborsRegressor()

# 검증하기
cv_score = cross_val_score(model_knn, X_train_s, y_train, cv=10, scoring='r2')

# 확인
print(cv_score)
print('평균: ', cv_score.mean())            # 모델의 평균 성능 평가(일반적인 성능을 대표하는 값)
print('표준편차: ', cv_score.std())          # 모델 성능의 일관성 평가(표준편차가 작다면 -> 모델이 안정적)

# 성능 정보 수집
result['KNN'] = cv_score.mean()
```

**3) Decision Tree**

```python
# 불러오기
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

# 선언하기
model_dt = DecisionTreeRegressor(max_depth=5, random_state=1)

# 검증하기
cv_score = cross_val_score(model_dt, X_train, y_train, cv=10, scoring='r2')

# 확인
print(cv_score)
print('평균: ', cv_score.mean())            # 모델의 평균 성능 평가(일반적인 성능을 대표하는 값)
print('표준편차: ', cv_score.std())          # 모델 성능의 일관성 평가(표준편차가 작다면 -> 모델이 안정적)

# 성능 정보 수집
result['Decision Tree'] = cv_score.mean()
```

**4) SVM**

```python
# 불러오기
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score

# 선언하기
model_svr = SVR()

# 검증하기
cv_score = cross_val_score(model_svr, X_train_s, y_train, cv=10, scoring='r2')

# 확인
print(cv_score)
print('평균: ', cv_score.mean())            # 모델의 평균 성능 평가(일반적인 성능을 대표하는 값)
print('표준편차: ', cv_score.std())          # 모델 성능의 일관성 평가(표준편차가 작다면 -> 모델이 안정적)

# 성능 정보 수집
result['svr'] = cv_score.mean()
```

# 6. 실제 성능 비교

```python
from sklearn.metrics import mean_absolute_error, r2_score

model = DecisionTreeRegressor()
model.fit(X_train_s, y_train)
y_pred = model.predict(X_test_s)
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
