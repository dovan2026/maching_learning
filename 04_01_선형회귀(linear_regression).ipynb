```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

# 3.데이터 전처리

**1) x, y 분리**

```python
target = 'dist'
X = df.drop(target, axis=1)
y = df[target]
```

**2) 학습용, 평가용 데이터 분리**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2026)
```

# 4.모델링


```python
# 1. 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# 2. 선언하기
model = LinearRegression()
```

```python
# 3. 학습하기
model.fit(X_train, y_train)
```

```python
# 4. 예측하기
y_pred = model.predict(X_test)
```

```python
print('MAE :', mean_absolute_error(y_test, y_pred))
print('R2 :', r2_score(y_test, y_pred))
```

- 회귀 계수

```python
print(model.coef_)      #가중치(기울기)
print(model.intercept_) #편향(y절편)
```

예) $$\large Distance = 4.05\times Speed-18.77  $$

# 5.기타

```python
# 시각화
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.ylabel('Dist(ft)')
plt.show()
```

```python
a = model.coef_     # 가중치(기울기)
b = model.intercept_ # 절편(상수항)

# 회귀선(직선)을 그리기 위한 x 구간 (최솟값~최댓값)
speed = np.array([X_train.min(), X_train.max()])
dist = a * speed + b    # 선형 회귀 공식

plt.scatter(X_test, y_test)     # 실제값
plt.scatter(X_test, y_pred)        # 예측값
plt.plot(speed, dist, color='r') # 회귀선 - 붉은 직선으로 모델의 예측 패턴 표시
plt.show()
```
